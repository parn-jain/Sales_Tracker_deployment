```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
connection = mysql.connector.connect(
    host="localhost",
    database="shanaya",
    user="root",
    password="Pj@123456",
    port = '3306'
)
from datetime import datetime, timedelta
```

```python
with connection.cursor() as cursor:
        cursor.execute("SELECT * FROM users_attendancerecord;")
        data = cursor.fetchall()
data
```

```python
df = pd.DataFrame(data, columns=['ID', 'Date', 'Start Time', 'End Time', 'Status', 'User ID'])

# Function to convert timedelta to hours, minutes, seconds
def timedelta_to_str(td):
    if pd.isna(td) or td is None:
        return None
    total_seconds = int(td.total_seconds())
    hours = total_seconds // 3600
    minutes = (total_seconds % 3600) // 60
    seconds = total_seconds % 60
    return f"{hours:02}:{minutes:02}:{seconds:02}"

# Apply the function to 'Start Time' and 'End Time'
df['Start Time'] = df['Start Time'].apply(timedelta_to_str)
df['End Time'] = df['End Time'].apply(timedelta_to_str)

# Display the DataFrame
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the initial 7 days
start_date = df['Date'].min()
end_date = start_date + pd.Timedelta(days=6)
df_initial_7_days = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Count the number of Present, Absent, and Late for each day
attendance_counts = df_initial_7_days.groupby(['Date', 'Status']).size().unstack(fill_value=0)

# Create bar chart
fig, ax = plt.subplots(figsize=(12, 6))

# Plot each status with a different color
attendance_counts.plot(kind='bar', ax=ax, color=['#1f77b4', '#ff7f0e', '#2ca02c'], width=0.8, edgecolor='black')

# Customize the plot
ax.set_xlabel('Date')
ax.set_ylabel('Total Number of Employees')
ax.set_title('Employee Attendance for Initial 7 Days')
ax.set_xticklabels(attendance_counts.index.strftime('%Y-%m-%d'), rotation=45)
ax.legend(title='Status')

plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Ensure 'Start Time' and 'End Time' are timedelta
df['Start Time'] = pd.to_timedelta(df['Start Time'].astype(str))
df['End Time'] = pd.to_timedelta(df['End Time'].astype(str))

# Filter data for the initial 7 days
start_date = df['Date'].min()
end_date = start_date + pd.Timedelta(days=6)
df_initial_7_days = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Calculate working hours per record
df_initial_7_days['Working Hours'] = df_initial_7_days['End Time'] - df_initial_7_days['Start Time']

# Aggregate the number of Present, Absent, and Late and working hours per day
attendance_counts = df_initial_7_days.groupby(['Date', 'Status']).size().unstack(fill_value=0)
working_hours_per_day = df_initial_7_days.groupby('Date')['Working Hours'].sum()

# Create bar chart
fig, ax = plt.subplots(figsize=(14, 7))

# Plot each status with a different color
attendance_counts.plot(kind='bar', ax=ax, color=['#1f77b4', '#ff7f0e', '#2ca02c'], width=0.8, edgecolor='black')

# Add KPI for working hours
for date, total_hours in working_hours_per_day.items():
    # Convert total_hours from timedelta to hours
    total_hours_hours = total_hours.total_seconds() / 3600
    ax.text(date, attendance_counts.loc[date].max() + 1, f"{total_hours_hours:.2f} hrs", 
            ha='center', va='bottom', fontsize=10, color='black')

# Customize the plot
ax.set_xlabel('Date')
ax.set_ylabel('Total Number of Employees')
ax.set_title('Employee Attendance and Working Hours for Initial 7 Days')
ax.set_xticklabels(attendance_counts.index.strftime('%Y-%m-%d'), rotation=45)
ax.legend(title='Status')

plt.tight_layout()
plt.show()
```
